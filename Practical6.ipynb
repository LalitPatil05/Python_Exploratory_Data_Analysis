```python
# Validating and Removing data manipulations (summarizing, sorting, sub-setting, metging, joining)

from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)

#outlier removal
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
df1 = df[~((df < (Q1 - 1.5*IQR)) | (df > (Q3 + 1.5*IQR))).any(axis=1)]

# Sorting
df = df.sort_values(by=df.columns[0])

# Filtering
print(df[df[df.columns[0]] > 5])

# Selecting
print(df[df.columns[1]])

# visualization
df.hist()
plt.show()
sns.boxplot(data=df)
```
