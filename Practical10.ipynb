```python
from sklearn.datasets import load_iris
import pandas as pd
from scipy.stats import chi2_contingency
import matplotlib.pyplot as plt
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
#preprocess
df=df.dropna()
#outlier removal
Q1=df.quantile(0.25)
Q3=df.quantile(0.75)
IQR=Q3-Q1
df=df[~((df < (Q1-1.5*IQR))|( df > (Q3+1.5*IQR))).any(axis=1)]
# Convert two features into categories (bins)
f1 = pd.cut(df[df.columns[0]], bins=3)
f2 = pd.cut(df[df.columns[1]], bins=3)
# Create contingency table
table = pd.crosstab(f1, f2)
# Chi-square test
chi2, p, dof, exp = chi2_contingency(table)
print("Chi2:", chi2)
print("P:", p)
#visualization
df.hist()
plt.show()
```
