```python
# Frequency Distribution

from sklearn.datasets import load_iris
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data=load_iris()
df=pd.DataFrame(data.data,columns=data.feature_names)

# Preprocessing
df = df.dropna()
#outlier removal
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
df1 = df[~((df < (Q1 - 1.5*IQR)) | (df > (Q3 + 1.5*IQR))).any(axis=1)]

# Choose one feature
col = df[df.columns[0]]

# Create bins (groups)
bins = pd.cut(col, bins=5)

# Frequency distribution
print(bins.value_counts())
df.hist()
plt.show()
sns.boxplot(data=df)
```
